```python
import numpy as np
```

```python
# Array creation
my_arr = np.array([1,2,3,4,5])   # list to numpy array
print(my_arr)
print(type(my_arr))
```

```python
# Different way to create arrays 

# 1. Array of zeros (useful for initializing)
zeros = np.zeros(5)
print("Zeros:", zeros)  

# 2. Array of ones
ones = np.ones(5)
print("Ones:", ones)  

# 3. Range of numbers (like Python's range)
range_array = np.arange(0, 10, 2)  # start, stop, step
print("Range:", range_array)  

# 4. Evenly spaced numbers
linspace_array = np.linspace(0, 1, 5)  # start, stop, how many numbers
print("Linspace:", linspace_array)  
```

```python
# Array attributes

arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)
print("Shape:", arr.shape)      # (5,) - means 5 elements in 1D
print("Data type:", arr.dtype)  # int64 - type of data stored
print("Number of dimensions:", arr.ndim)  # 1 - one-dimensional
print("Size (total elements):", arr.size)  # 5
```
