```python
# Array slicing 

import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60, 70])

# Slicing syntax: arr[start:stop:step]
print("First 3:", arr[0:3])      # [10 20 30]
print("From index 2 to 5:", arr[2:5])  # [30 40 50]
print("Every other element:", arr[::2]) # [10 30 50 70]
print("First 4:", arr[:4])       # [10 20 30 40] (start defaults to 0)
print("Last 3:", arr[-3:])       # [50 60 70]
print("Reverse:", arr[::-1])     # [70 60 50 40 30 20 10]
```

```python
# Array modification

arr = np.array([10, 20, 30, 40, 50])

# Modify single element
arr[0] = 100
print("After modifying first:", arr)  # [100  20  30  40  50]

# Modify multiple elements using slicing
arr[1:4] = 999
print("After slice modification:", arr)  # [100 999 999 999  50]

# Modify using another array
arr[1:4] = np.array([200, 300, 400])
print("After array modification:", arr)  # [100 200 300 400  50]
```

```python
# Basic arithmetic operations

arr = np.array([1, 2, 3, 4, 5])

# Element-wise operations
print("Add 10:", arr + 10)        # [11 12 13 14 15]
print("Multiply by 2:", arr * 2)  # [ 2  4  6  8 10]
print("Square:", arr ** 2)        # [ 1  4  9 16 25]
print("Divide by 2:", arr / 2)    # [0.5 1.  1.5 2.  2.5]
```

```python
# Array to arrya operations

import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([10, 20, 30, 40])

# Element-wise operations between arrays
# array must have same shape
print("Addition:", arr1 + arr2)       # [11 22 33 44]
print("Multiplication:", arr1 * arr2) # [10 40 90 160]
print("Subtraction:", arr1 - arr2)    # [-9 -18 -27 -36]
```

```python
# Feature normalization

# Raw feature values (e.g., house prices in dollars)
prices = np.array([300000, 450000, 200000, 800000, 350000])

# Normalization: scale to 0-1 range
min_price = prices.min()
max_price = prices.max()
normalized = (prices - min_price) / (max_price - min_price)

print("Original:", prices)
print("Normalized:", normalized)
```
