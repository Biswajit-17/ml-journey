```python
# Sample Data

import pandas as pd
import numpy as np

# Create a realistic e-commerce dataset
np.random.seed(42)

data = {
    'order_id': range(1, 21),
    'customer_age': [25, 34, 45, 23, 56, 38, 29, 41, 33, 50, 
                     27, 44, 36, 52, 31, 48, 26, 39, 42, 35],
    'product_price': [29.99, 149.99, 79.99, 19.99, 299.99, 
                      89.99, 49.99, 199.99, 59.99, 249.99,
                      39.99, 169.99, 99.99, 189.99, 69.99,
                      219.99, 44.99, 129.99, 159.99, 109.99],
    'quantity': [1, 2, 1, 3, 1, 2, 1, 1, 2, 1,
                 3, 1, 2, 1, 1, 1, 2, 1, 1, 2],
    'category': ['Electronics', 'Electronics', 'Clothing', 'Books', 'Electronics',
                 'Clothing', 'Books', 'Electronics', 'Clothing', 'Electronics',
                 'Books', 'Electronics', 'Clothing', 'Electronics', 'Clothing',
                 'Electronics', 'Books', 'Electronics', 'Electronics', 'Clothing'],
    'customer_type': ['New', 'Returning', 'Returning', 'New', 'VIP',
                      'Returning', 'New', 'VIP', 'Returning', 'VIP',
                      'New', 'Returning', 'Returning', 'VIP', 'New',
                      'VIP', 'New', 'Returning', 'VIP', 'Returning']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Single conditions

only_books = df[df['category'] == 'Books']
print(only_books)

quantity_not_one = df[df['quantity'] != 1]
print(quantity_not_one)
```

```python
# Multiple conditions

high_returning = df[(df['product_price'] > 100) & (df['customer_type'] == 'Returning')]
print(high_returning)

high_vip = df[(df['product_price'] > 100) | (df['customer_type'] == 'Returning')]
print(high_vip)
```

```python
# Filter for multiple specific values
categories_of_interest = ['Electronics', 'Books']
filtered = df[df['category'].isin(categories_of_interest)]
print("ELECTRONICS AND BOOKS ONLY:")
print(filtered)

# Opposite: NOT in list
not_clothing = df[~df['category'].isin(['Clothing'])]
print("EVERYTHING EXCEPT CLOTHING:")
print(not_clothing)
```

```python
# Filter for VIP customers (string contains)
vip = df[df['customer_type'].str.contains('VIP')]
print("VIP CUSTOMERS:")
print(vip)

print("\n" + "="*60 + "\n")

# Filter for categories starting with 'E'
e_categories = df[df['category'].str.startswith('E')]
print("CATEGORIES STARTING WITH 'E':")
print(e_categories)
```

```python
# Bonus

# Create a new dataset
np.random.seed(100)
ml_data = {
    'customer_id': range(1, 101),
    'age': np.random.randint(18, 75, 100),
    'income': np.random.randint(20000, 150000, 100),
    'purchase_amount': np.random.randint(10, 1000, 100),
    'num_purchases': np.random.randint(1, 50, 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'subscription': np.random.choice(['Basic', 'Premium', 'Enterprise'], 100)
}

df_ml = pd.DataFrame(ml_data)
print(df_ml.head())
print(df_ml.shape)
```

```python
# 1. Select features for modeling: age, income, purchase_amount, num_purchases
features = df_ml[['age', 'income', 'purchase_amount', 'num_purchases']]
print(features.head())
```

```python
# 2 & 3. Filter: Keep only customers aged 25-60 and purchase_amount between 50 and 800
filtered_df = df_ml[(df_ml['age'] >=25) & (df_ml['age'] <=60) & (df_ml['purchase_amount'] >= 50) & (df_ml['purchase_amount'] <= 800)]
print(filtered_df.head())
```

```python
# 4. Create 3 separate DataFrames by region: North, South, and East+West combined
north_df = df_ml[df_ml['region'] == 'North']
south_df = df_ml[df_ml['region'] == 'South']
east_west_df = df_ml[(df_ml['region'] == 'East') | (df_ml['region'] == 'West')]

print("North: \n", north_df.head(3))
print("South: \n", south_df.head(3))
print("East and West:", east_west_df.head(3))
```
