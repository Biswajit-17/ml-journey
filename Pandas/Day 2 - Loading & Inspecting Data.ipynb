```python
import pandas as pd
import numpy as np

# Create a realistic customer dataset
np.random.seed(42)

data = {
    'customer_id': range(1, 101),
    'age': np.random.randint(18, 70, 100),
    'income': np.random.randint(20000, 150000, 100),
    'credit_score': np.random.randint(300, 850, 100),
    'loan_amount': np.random.randint(5000, 50000, 100),
    'employment_years': np.random.randint(0, 30, 100),
    'defaulted': np.random.choice([0, 1], 100, p=[0.8, 0.2])
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.head())
print(df.head(10))
print(df.tail())
```

```python
print(df.info())  # Structure of data
```

```python
print(df.describe())  # Statistical summary
```

```python
print(df.columns.tolist())
print(df.dtypes)
print("Nulls:", df.isnull().sum())
```
