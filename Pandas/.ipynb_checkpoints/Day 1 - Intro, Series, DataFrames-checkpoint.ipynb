```python
import pandas as pd
```

```python
ages = pd.Series([18,19,20,21])
print(ages)
```

```python
# Series with meaningful labels
ages = pd.Series([25, 30, 35, 40, 45], 
                 index=['Alice', 'Bob', 'Charlie', 'David', 'Eve'])
print(ages)
```

```python
data = {
    "name" : ["bob", "alex", "jack", "mat"],
    "age" : [18, 19, 20, 19],
    "salary" : [50000, 60000, 70000, 80000],
    "experience" : [2, 5, 8, 12]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Accessing columns (returns a Series)
print("Ages column:")
print(df['age'])

print("\n" + "="*40)

# Accessing multiple columns (returns a DataFrame)
print("Age and Salary:")
print(df[['age', 'salary']])

print("\n" + "="*40)

# Basic info about DataFrame
print("Shape (rows, columns):", df.shape)
print("Column names:", df.columns.tolist())
print("Data types:\n", df.dtypes)
```

```python
# Simulating ML feature data
ml_data = {
    'feature_1': [1.2, 2.3, 3.1, 4.5, 5.0],
    'feature_2': [10, 20, 15, 25, 30],
    'feature_3': [0.5, 0.6, 0.7, 0.8, 0.9],
    'target': [0, 1, 0, 1, 1]  # What we want to predict
}

ml_df = pd.DataFrame(ml_data)
print(ml_df)

# Separate features from target (common ML step)
X = ml_df[['feature_1', 'feature_2', 'feature_3']]  # Features
y = ml_df['target']  # Target

print("\nFeatures (X):")
print(X)
print("\nTarget (y):")
print(y)
```
